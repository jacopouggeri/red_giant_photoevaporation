```python
# Modules
import numpy as np
import matplotlib.pyplot as plt # for plotting
from EUV_flux import euv_flux_photons
%matplotlib inline
from astropy.constants import M_sun

# Define matplotlibve style 
plt.style.use("defstyle.mplstyle")

# Ignore divide by zero errors (will generate NaNs but those will be replaced by zeros)
np.seterr(divide='ignore')
```

```python
# Black hole parameters
M = 4e6 * M_sun
m_dot = np.logspace(-3, 0, 10)

# Calculate different photon fluxes
photon_fluxes = []
for el in m_dot:
    photon_fluxes.append(euv_flux_photons(el, M).to("Hz").value)

# Generate plot of photon flux as function of accretion rate
plt.loglog(m_dot, photon_fluxes)
plt.xlabel(r"$\frac{\dot M}{\dot M_{edd}}$")
plt.ylabel(r"$\Phi_{\gamma} [s^{-1}]$")
```
